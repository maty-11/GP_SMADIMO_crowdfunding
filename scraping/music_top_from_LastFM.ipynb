```python
tracks_with_info = []
for page in range(1, 210):
    params = {"method": "chart.getTopTracks",
        "api_key": API_key,
        "format": "json",
        "page": page}
    toplist = requests.get("https://ws.audioscrobbler.com/2.0/", params=params)
    res = toplist.json()
    tracks = res["tracks"]["track"]

    for track in tracks:
      tracks_with_info.append({"name": track["name"], "artist": track["artist"]["name"],
                               "playcount": track["playcount"], "listeners": track["listeners"], "duration": track["duration"],"url": track["url"]})
df = pd.DataFrame(tracks_with_info)
print(df)
df.to_csv('output.csv', index=False)
```

```python
tracks_with_info = []
for page in range(1, 210):
    params = {"method": "chart.getTopTracks",
        "api_key": API_key,
        "format": "json",
        "page": page}
    toplist = requests.get("https://ws.audioscrobbler.com/2.0/", params=params)
    res = toplist.json()
    tracks = res["tracks"]["track"]

    for track in tracks:
      tracks_with_info.append({"name": track["name"], "artist": track["artist"]["name"],
                                "playcount": track["playcount"], "listeners": track["listeners"], "duration": track["duration"],"url": track["url"]})
      iparams = {
          "method": "track.getInfo",
          "api_key": API_key,
          "artist":  track["artist"]["name"],
          "track":track["name"],
          "format": "json"}
      itoplist = requests.get("https://ws.audioscrobbler.com/2.0/", params=iparams)
      ires = itoplist.json()
      if 'error' not in ires:
        tags = ires['track'].get("toptags", {}).get("tag", [])
        genres = []
        if type(tags) == list:
          for tag in tags:
            genres.append(tag["name"])
        tracks_with_info[-1]["genres"] =  str(";".join(genres))
        if genres:
          tracks_with_info[-1]["top_genre"] = genres[0]


df = pd.DataFrame(tracks_with_info)
print(df.info())
df.to_excel('output.xlsx')
```
