Делаем базовый запрос к списку активных проектов

```python
import requests
import pandas as pd
import time
```

```python
base_url = 'https://www.indiegogo.com'
page = requests.get(f'{base_url}/api/public/projects/getActiveCrowdfundingProjects')
page
```

```python
res = page.json()
res[:2]
```

```python
projects_df = pd.DataFrame(res)
projects_df
```

Преобразуем даты старта и окончания кампании, для фьючер инжиниринга

```python
projects_df['campaignStartDate'] = pd.to_datetime(projects_df['campaignStartDate'], errors='coerce')
projects_df['campaignEndDate'] = pd.to_datetime(projects_df['campaignEndDate'], errors='coerce')

projects_df[['campaignStartDate', 'campaignEndDate']].head()
```

Добавили новые признаки: процент выполнения цели, длительность кампании и бинарный признак “цель достигнута или нет”.

```python
projects_df['pct_funded'] = projects_df['fundsGathered'] / projects_df['campaignGoal']
projects_df['days_campaign'] = (projects_df['campaignEndDate'] - projects_df['campaignStartDate']).dt.days
projects_df['is_overfunded'] = (projects_df['fundsGathered'] >= projects_df['campaignGoal']).astype(int)

projects_df[['projectName', 'campaignGoal', 'fundsGathered', 'pct_funded', 'days_campaign', 'is_overfunded']].head()
```

По этим идентификаторам потом можно запрашивать детали по каждому проекту и автору отдельно.

```python
project_url_names = projects_df['projectUrlName'].dropna().unique()
creator_url_names = projects_df['creatorUrlName'].dropna().unique()
```

Пишем функцию для запроса одного проекта по urlName

```python
def get_project_by_url_name(url_name):
    params = {'urlName': url_name}
    page = requests.get(f'{base_url}/api/public/projects/getCrowdfundingProject', params=params)
    return page

page = get_project_by_url_name(project_url_names[0])
page
```

```python
project_example = page.json()
project_example
```

Делаем цикл и собираем данные по каждому проекту

```python
projects_full = []

for url_name in project_url_names:
    page = get_project_by_url_name(url_name)

    if page.status_code == 200:
        res = page.json()
        projects_full.append(res)

    time.sleep(0.2)
```

```python
projects_full_df = pd.DataFrame(projects_full)
projects_full_df.head()
```

Пишем функцию для получения данных автора

```python
def get_creator_by_url_name(url_name):
    params = {'urlName': url_name}
    page = requests.get(f'{base_url}/api/public/creators/getCreator', params=params)
    return page

page = get_creator_by_url_name(creator_url_names[0])
page
```

```python
creator_example = page.json()
creator_example
```

```python
creators_full = []

for url_name in creator_url_names:
    page = get_creator_by_url_name(url_name)

    if page.status_code == 200:
        res = page.json()
        creators_full.append(res)

    time.sleep(0.2)
```

```python
creators_df = pd.DataFrame(creators_full)
creators_df.head()
```
